```python
# Import pandas library, commonly used for data manipulation and analysis
import pandas as pd 
```

```python
# Load the salary dataset from a CSV file into a pandas DataFrame
dataset= pd.read_csv("Salary_Data.csv")
```

```python
dataset
```

```python
independent = dataset[["YearsExperience"]]
```

```python
independent
```

```python
dependent = dataset[["Salary"]]
```

```python
dependent
```

```python
# Split the dataset into training and testing sets
# independent variables (x) and dependent variables (y) with 30% of data used for testing
# random_state=0 ensures reproducibility of the split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
# Import the LinearRegression model from scikit-learn
from sklearn.linear_model import LinearRegression

# Create a LinearRegression model instance and create a variable name :
regressor = LinearRegression()

# Train the model using the training data (x_train and y_train)
regressor.fit (x_train,y_train)
```

```python
# Use the trained regressor model to make predictions on the test data:
y_pred = regressor.predict(x_test)
```

```python
# Import the r2_score function from sklearn.metrics to evaluate regression model performance
from sklearn.metrics import r2_score
# Calculate the R-squared score by comparing the actual test values (y_test) with the predicted values (y_pred)
# R-squared measures how well the model explains the variance in the target variable (1.0 is perfect prediction)
r_score = r2_score(y_test,y_pred)
```

```python
# Accessing the r_score variable, likely containing the R-squared score from a regression model:
r_score
```

```python
# Import the pickle module for object serialization/deserialization:
import pickle

# Define the filename for the saved model
# This will be used to load or save a trained linear regression model
filename = "finalized_model_linear.sav"
```

```python
# Save the trained regressor model to a file using pickle
# This serializes the model object and writes it to disk in binary format
pickle.dump(regressor,open(filename,"wb"))
```

```python
# Load the pre-trained linear regression model from a saved file:
load_model = pickle.load(open("finalized_model_linear.sav","rb"))
# Use the loaded model to predict the output for input value 13:
result = load_model.predict([[11]])
```
